```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_excel('./data/churn.xls', sheet_name='churn', index_col='Phone')
```

```python
df.head()
```

**Goal:** Get some basic statistics at the State/Area Code level.

```python
df.groupby('State').mean() # Average per State of each column
```

```python
df.columns
```

```python
df.groupby('State')[['Day Mins', 'Day Charge']].mean()
```

```python
df.groupby('State')[['Day Mins', 'Day Charge']].agg({'Day Mins':'sum', 'Day Charge':
                                                    'mean'}) # Apply different operations to grps
```

```python
# Filter + Grouping + Aggregate
df[df['State'].isin(['AK','AL'])].groupby(['State', 'Churn'])['Day Charge'].mean()
```

## Pivot tables

```python
df.pivot_table(columns='State', index='Churn', values='Day Charge', aggfunc='mean')
```
